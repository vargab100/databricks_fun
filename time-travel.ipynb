# **add new data**

```python
from pyspark.sql import Row

# New sample row
new_row = [Row(
    tpep_pickup_datetime="2025-11-10 10:00:00",
    tpep_dropoff_datetime="2025-11-10 10:20:00",
    trip_distance=3.2,
    fare_amount=15.5,
    pickup_zip="10005",
    dropoff_zip="10006"
)]

df_new = spark.createDataFrame(new_row)

# Append new row to Delta table
df_new.write.format("delta").mode("append").saveAsTable("workspace.my_project_dev.demo_table")

display(spark.table("workspace.my_project_dev.demo_table"))
```

# **check the describe history **

```python
%sql
describe history workspace.my_project_dev.demo_table
```

# **Call the specific version**

```python
# Read Delta table at a specific version
df_version = spark.read.format("delta").option("versionAsOf", 1).table("workspace.my_project_dev.demo_table")
display(df_version)
```
